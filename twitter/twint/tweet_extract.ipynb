```python
import pandas as pd
raw = pd.read_csv('ts_data/ts.csv')
```

```python
raw.columns
```

```python
raw.drop(labels='cashtags', axis=1, inplace= True)
```

```python
raw.drop(labels='translate', axis=1, inplace= True)
raw.drop(labels='trans_src', axis=1, inplace= True)
raw.drop(labels='trans_dest', axis=1, inplace= True)
raw.drop(labels='place', axis=1, inplace= True)
raw.drop(labels='thumbnail', axis=1, inplace= True)
```

```python
raw.columns
```

```python
twt = raw[['id', 'date','time', 'tweet','language']]
twt.to_csv('ts_data/tweet.csv')
```

```python
user = raw[['user_id','username','name']]
user.drop_duplicates('user_id', inplace = True)
user.reset_index(inplace = True)
user.drop(labels='index', axis=1, inplace = True)
user.to_csv('ts_data/user.csv')
# user.to_csv('')
```

```python
data = {'tweetId':[],'hashtag':[]}

for index, row in raw.iterrows():

    # print(row['id'])

    if row.hashtags != '[]':
        tmp = row.hashtags[1:-1]
        tmp = tmp.split(',')
        for h in tmp:
            h = h.strip()[1:-1]
            print(row.id, '\t', h)
            data['tweetId'].append(row.id)
            data['hashtag'].append(h)

htag = pd.DataFrame(data)            

htag.to_csv('ts_data/hashtags.csv')
```
